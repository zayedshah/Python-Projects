```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('http://ft.com').text
soup = BeautifulSoup(source,'lxml')
```

```python
print('Section', '\t\t\t\t', 'Headline')
print('*' * 100)
for t in soup.find_all('div',{'class':'o-teaser__content'}):
    try:
        section = t.find('a',{'class':'o-teaser__tag'}).text.strip()
    except:
        section = ''
    
    try:
        headline = t.find('a',{'class':'js-teaser-heading-link'}).text.strip()
    except:    
        headline = None
    print(section, '\n\t\t\t\t\t', headline)
```
